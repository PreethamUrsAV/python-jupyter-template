### Exercise 5: Subplot Layout

```python

import matplotlib.pyplot as plt
import numpy as np

# Data for line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Data for bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

# Data for scatter plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

# Data for pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line plot
axs[0, 0].plot(x, y, color='b', linewidth=2)
axs[0, 0].set_title('Line Chart')
axs[0, 0].grid(True, linestyle='--', alpha=0.6)

# Bar plot
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title('Bar Chart')

# Scatter plot
axs[1, 0].scatter(x_scatter, y_scatter, color='blue', marker='o')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].grid(True, linestyle='--', alpha=0.6)

# Pie chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'], startangle=140, wedgeprops={'edgecolor': 'black'})
axs[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()
plt.show()
```
